```python
import sys
assert sys.version_info >= (3,5)

#is this notebook running on Colab or Kaggle? # 코랩에서 실행되냐?
IS_COLAB = "google.colab" in sys.modules

#Scikit-learn >= 0.20 is required # 사이킷런 버전 설정
import sklearn
assert sklearn.__version__ >= "0.20"

#Common imports #당연히 임포트 하는 것들 
import numpy as np
import os

#to make this notebook's output stable across runs # ??
np.random.seed(42)

# To plot pretty figures # matplotlib 설정?
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

#Where to save the figures # 어디에 저장할지?
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "classification"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

#png로 저장하는 함수
def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
    

def display_scores(scores):
    print("scores :", scores)
    print("mean :", scores.mean())
    print("standard deviation :", scores.std())
```

```python
import pandas as pd

DATA_PATH = "datasets"

def load_data():
    csv_path = os.path.join(DATA_PATH, "kc_house_data.csv")
    return pd.read_csv(csv_path) 
```

### 특성 조합 테스트

```python
drop_set01 = ["condition", "long", "id", "date"]

drop_set02 = ["id", "date"]

drop_set03 = ["date"]

drop_set04 = []
```

```python
new_pd_train = pd_train.drop(drop_set04, axis=1)

new_pd_train.keys()
```

```python
new_pd_train["latest spacious house that expect expensive"] = (new_pd_train["sqft_living"] * new_pd_train["floors"] ) / ( new_pd_train["yr_built"])

new_pd_train.plot(kind = "scatter", y = "price", x = "latest spacious house that expect expensive", alpha = 0.2)
plt.show()
```

```python
for i in range(len(new_pd_train)):
    if new_pd_train["view"][i] == 0:
        print(i)
    new_pd_train["date"][i] = int(new_pd_train["date"][i][0:8])
        
# view 특성에 0값이 많음 확인

# date 특성의 의미없는 부분 제거
```

```python
new_pd_train["date"][0]
```

```python
new_pd_train["overall value"] = new_pd_train["grade"] * ( new_pd_train["view"] + new_pd_train["condition"] )

new_pd_train.plot(kind = "scatter", y = "price", x = "overall value", alpha = 0.2)
plt.show()
```

```python
new_X = new_pd_train.drop(['price'],axis=1).values
new_y = new_pd_train['price'].values
```

```python
print(new_X.shape)
print(new_y.shape)
```

```python
import numpy as np
shuffle_index = np.random.permutation(21613)
new_X_train, new_y_train = new_X[shuffle_index], new_y[shuffle_index]
```

```python
corr_matrix = new_pd_train.corr()

corr_matrix["price"].sort_values(ascending=False)
```

#### Random forest test

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor

forest_clf = RandomForestRegressor(random_state=42)
```

```python
forest_mse = cross_val_score(forest_clf, new_X_train, new_y_train, cv=10, scoring="neg_mean_squared_error")
```

```python
forest_rmse = np.sqrt(-forest_mse)
display_scores(forest_rmse)
```

#### SVR test

```python
from sklearn.svm import SVR

svr01 = SVR(kernel='linear', C=10, gamma='auto')
svr02 = SVR(kernel='poly', C=10, gamma='auto')
svr03 = SVR(kernel='sigmoid', C=10, gamma='auto')
```

```python
svr_mse01 = cross_val_score(svr01, new_X_train, new_y_train, cv=10, scoring="neg_mean_squared_error")
```
